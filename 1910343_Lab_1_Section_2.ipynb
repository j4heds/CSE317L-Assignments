##### **Problem 01:**
Calculate the sum and mean of the 2nd row of the matrix.

```python
import numpy as np

matrix = np.random.rand(10, 5)  # Assuming you have a 10x5 matrix

second_row = matrix[1, :]  # Extract the 2nd row from the matrix

row_sum = np.sum(second_row)
row_mean = np.mean(second_row)

print("Sum of the 2nd row:", row_sum)
print("Mean of the 2nd row:", row_mean)
```

##### **Problem 04:**
Generate the vector `[25, 16, 9, 4, 1]` and then find the square root of each element of the vector using for loop. **Hint:** Use `np.sqrt()` function to find the square root of a number, don't use any other built-in function.

```python
import numpy as np

vector = np.array([25, 16, 9, 4, 1])
result = np.zeros_like(vector)  # Create an array of zeros with the same shape as the vector

for i in range(len(vector)):
    result[i] = np.sqrt(vector[i])  # Compute the square root of each element

print("Square roots:", result)
```

##### **Problem 05:**
Write a function that takes a number and returns a vector like the previous problem. For example, if the input is 5, then the output should be `[25, 16, 9, 4, 1]`. If the input is 10, then the output should be `[100, 81, 64, 49, 36, 25, 16, 9, 4, 1]`.

```python
import numpy as np

def generate_square_vector(n):
    vector = np.arange(n, 0, -1)  # Create a vector from n to 1 in descending order
    vector = vector ** 2  # Compute the squares of each element
    return vector

# Test the function
n = 5
result = generate_square_vector(n)
print("Output for n =", n, ":", result)

n = 10
result = generate_square_vector(n)
print("Output for n =", n, ":", result)
```

##### **Problem 08:**
Plot $x + y + z = 1$ for $x, y \in [-3, 3]$ with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the grid
x = np.arange(-3, 3.5, 0.5)  # x values from -3 to 3 with interval 0.5
y = np.arange(-3, 3.5, 0.5)  # y values from -3 to 3 with interval 0.5
X, Y = np.meshgrid(x, y)  # Create a grid of x and y values

# Compute the corresponding z values
Z = 1 - X - Y

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Surface Plot: x + y + z = 1')

# Show the plot
plt.show()
```

Download the image of Lena Forsen from the following link: [https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png](https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png). Then, do the following tasks:

*  Load the image to a numpy array.
*  Find the size of the image.
*  Find the minimum and maximum value of the image.
*  Find the mean and standard deviation of the image.
*  Plot the image using `plt.imshow()` or `display()`.
*  Resize the image to 50 x 50.
*  Convert the image to grayscale.
*  Plot the grayscale image using `plt.imshow()` or `display()`.
*  Save the grayscale image using `imsave()`.
*  Stack the columns of the image to form a vector of size 2500 x 1.

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color

# Load the image to a numpy array
image_url = 'https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png'
image = io.imread(image_url)

# Find the size of the image
image_size = image.shape
print("Image size:", image_size)

# Find the minimum and maximum value of the image
image_min = np.min(image)
image_max = np.max(image)
print("Minimum value:", image_min)
print("Maximum value:", image_max)

# Find the mean and standard deviation of the image
image_mean = np.mean(image)
image_std = np.std(image)
print("Mean:", image_mean)
print("Standard deviation:", image_std)

# Plot the image
plt.imshow(image)
plt.axis('off')
plt.title('Original Image')
plt.show()

# Resize the image to 50 x 50
resized_image = io.resize(image, (50, 50), anti_aliasing=True)

# Convert the image to grayscale
gray_image = color.rgb2gray(resized_image)

# Plot the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title('Grayscale Image')
plt.show()

# Save the grayscale image
io.imsave('grayscale_image.png', gray_image)

# Stack the columns of the image to form a vector of size 2500 x 1
image_vector = np.reshape(resized_image, (-1, 1))
print("Image vector shape:", image_vector.shape)
```

##### **Problem 10:**
Compute 2nd order Taylor approximation of $e^{2x}$ and $e^{3x}$, and $e^{5x}$ around $a = 0$ for $x \in [-1, 1]$ with interval 0.01. Then, determine which of the following is more accurate:
> 1. Calculating the Taylor approximation separately for each $e^{2x}$, $e^{3x}$ then multiplying them together.
> 2. Calculating the Taylor approximation for $e^{5x}$.

Use mean square error (MSE) to determine the accuracy. Also, plot the Taylor approximation for 1st and 2nd method and the original function in the same plot for visual comparison.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions and their derivatives
def exponential_2x(x):
    return np.exp(2 * x)

def exponential_3x(x):
    return np.exp(3 * x)

def exponential_5x(x):
    return np.exp(5 * x)

def taylor_approximation_2nd_order(f, a, x):
    f_a = f(a)
    f_a_prime = f(a) * 2
    f_a_double_prime = f(a) * 2**2

    return f_a + f_a_prime * (x - a) + 0.5 * f_a_double_prime * (x - a)**2

# Define the interval and generate x values
interval = np.arange(-1, 1.01, 0.01)

# Compute the 2nd order Taylor approximations for each function
taylor_approximation_2x = taylor_approximation_2nd_order(exponential_2x, 0, interval)
taylor_approximation_3x = taylor_approximation_2nd_order(exponential_3x, 0, interval)
taylor_approximation_5x = taylor_approximation_2nd_order(exponential_5x, 0, interval)

# Calculate the actual function values
actual_2x = exponential_2x(interval)
actual_3x = exponential_3x(interval)
actual_5x = exponential_5x(interval)

# Calculate the mean square error (MSE) for each approximation
mse_1 = np.mean((taylor_approximation_2x * taylor_approximation_3x - actual_5x)**2)
mse_2 = np.mean((taylor_approximation_5x - actual_5x)**2)

print("MSE for multiplying two approximations:", mse_1)
print("MSE for single approximation:", mse_2)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(interval, taylor_approximation_2x * taylor_approximation_3x, label="Method 1: Multiplying approximations")
plt.plot(interval, taylor_approximation_5x, label="Method 2: Single approximation")
plt.plot(interval, actual_5x, label="Actual function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Comparison of Taylor Approximations")
plt.legend()
plt.show()
```
